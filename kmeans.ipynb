```python
#Imports
import pandas as pd
import numpy as np
iris = pd.read_csv("iris.csv")
iris.head()
```

```python
#Análise da Base
iris.info()
```

```python
#Remover os Atributo Id & Species 
X = iris.iloc[:, 1:4].values
X
```

#  

# Método Elbow

```python
#O método Elbow se trata de uma técnica interessante para encontrar o valor ideal do parâmetro k.
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
wcss = []
 
for i in range(1, 11):
    kmeans = KMeans(n_clusters = i, init = 'random')
    kmeans.fit(X)
    print(i,kmeans.inertia_)
    wcss.append(kmeans.inertia_)  
plt.plot(range(1, 11), wcss)
plt.title('O Metodo Elbow')
plt.xlabel('Numero de Clusters')
plt.ylabel('WSS') #A soma dos quadrados dentro do cluster é uma medida da variabilidade das observações dentro de cada cluster.
plt.show()
```

```python
#Algoritmo K-means e seus parâmetros
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters = 3, init = 'random')
```

```python
#O método fit() recebe como parâmetro os dados a serem agrupados, nesse caso será a variável X que declaramos anteriormente.
kmeans.fit(X)
```

```python
#Neste momento já temos os dados agrupados e vamos verificar os centroides gerados através do atributo cluster_centers_.
kmeans.cluster_centers_[:,:]
```

```python
# Computa o centros de cluster e prever índice de cluster para cada amostra.
previsao = kmeans.fit_predict(X)
previsao
```

# Centroids

```python
plt.scatter(X[previsao == 0, 0], X[previsao == 0, 1], s = 50, c = 'red', label = 'Cluster 1')
plt.scatter(X[previsao == 1, 0], X[previsao == 1, 1], s = 50, c = 'blue', label = 'Cluster 2')
plt.scatter(X[previsao == 2, 0], X[previsao == 2, 1], s = 50, c = 'green', label = 'Cluster 3')
plt.scatter(X[previsao == 3, 0], X[previsao == 3, 1], s = 100, c = 'cyan', label = 'Cluster 4')
plt.scatter(X[previsao == 4, 0], X[previsao == 4, 1], s = 100, c = 'magenta', label = 'Cluster 5')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], s = 150, c = 'yellow', label = 'Centroids')
plt.title('Clusters of customers')
plt.legend()
plt.show()
```
